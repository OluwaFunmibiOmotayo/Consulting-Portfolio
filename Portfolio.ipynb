# Abstract

This study evaluates how intercropping peppers with flowering plants (Sunflower, Zinnia, Dianthus) compared to a pepper monoculture influences pepper yield, disease incidence, and insect population dynamics. Using a randomized complete block design, yield measurements were separately recorded for peppers near the intercrops and those farther away. Results reveal significant differences in yield outcomes and potential agronomic advantages of specific intercropping strategies.

# Keywords

Peppers; Intercropping; Flowering plants; Yield; Disease incidence; Insect populations

# Introduction

Intercropping has been increasingly recognized as a strategy to improve crop yield, enhance pest management, and promote sustainable agriculture. Several studies have shown that flowering plants can attract beneficial insects, potentially reducing pest pressures on nearby crops \cite{Altieri1999, Ratnadass2012}. However, limited research exists on how different flower species specifically influence pepper (Capsicum spp.) production.

In this study, we investigate the effects of intercropping peppers with Sunflower (Helianthus annuus), Zinnia (Zinnia elegans), and Dianthus (Dianthus spp.) compared to a monoculture control. The primary objectives were to:
- Assess changes in pepper yield,
- Evaluate disease incidence,
- Monitor insect populations across treatments.

# Methods

## Experimental Design

The experiment was conducted using a randomized complete block design. Each block contained four treatments:
1. Pepper monoculture (control),
2. Pepper intercropped with Sunflower,
3. Pepper intercropped with Zinnia,
4. Pepper intercropped with Dianthus.

Within each plot, peppers were categorized based on proximity to the flowering plants:
- "Inner" peppers (close to flowering plant),
- "Outer" peppers (farther away).

## Data Collection

- **Yield:** Measured as total fruit weight per plant at harvest.
- **Disease Incidence:** Recorded visually based on presence of symptoms.
- **Insect Populations:** Monitored using sticky traps and visual counts.

## Statistical Analysis

Data were analyzed using analysis of variance (ANOVA) models with treatment and position as fixed effects, and block as a random effect. Post-hoc comparisons were performed using Tukeyâ€™s HSD test. All analyses were conducted using R (version 4.3.0).

# Results

## Yield

Yield varied significantly across treatments (\(P < 0.05\)), with peppers intercropped with Zinnia showing the highest yields among all groups. Inner peppers generally outperformed outer peppers, suggesting a localized positive effect of intercropping.

## Disease Incidence

Disease incidence was lowest in peppers intercropped with Dianthus, though differences were not statistically significant.

## Insect Populations

Higher beneficial insect counts were recorded in Sunflower-intercropped plots, while monoculture plots had higher pest presence.

# Discussion

Our findings suggest that intercropping peppers with certain flowering species, particularly Zinnia, can enhance pepper yields. Potential mechanisms include increased pollinator activity, improved microclimate conditions, or pest suppression.

While disease incidence did not significantly vary, trends indicated potential benefits with Dianthus. Further research over multiple growing seasons is recommended to validate these observations and explore underlying biological interactions.

# Conclusion

Intercropping peppers with flowering plants offers promising agronomic benefits. Strategic species selection can optimize yield and contribute to sustainable pepper production systems.
